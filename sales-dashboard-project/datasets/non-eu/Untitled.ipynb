```python
import pandas as pd

# 使用 pandas 读取 CSV 文件
data = pd.read_csv('Orders.csv',delimiter=';')
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'], format='%d/%m/%Y')

# 提取年份并创建一个新的列 'Year'
data['Year'] = data['Order Date'].dt.year

# 按年份对 'Quantity' 列进行求和
quantity_sum_per_year = data.groupby('Year')['Quantity'].sum()
```

```python
print(quantity_sum_per_year)
```
